```python
X = df.iloc[:, :7].values
```

```python
y = df.iloc[:, 7].values
```

```python
print(pd.DataFrame(X[:10]))
print()
print(pd.DataFrame(y[:10]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()
classifier.add(Dense(activation="relu", input_dim=7, units=4, kernel_initializer="uniform"))
classifier.add(Dense(activation="relu", units=7, kernel_initializer="uniform"))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 25, epochs = 200)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
tn, fp, fn, tp = cm.ravel()
accuracy = (tn + tp) / (tn + tp + fn + fp)
print(accuracy)
```

According to the confusion matrix, the model correctly identified positives and negatives approximately 69% of the time. This means that the model has an accuracy of about 69%.

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
accuracies = []
for k in range(1, 65):
    classifier = KNeighborsClassifier(n_neighbors= k)
    classifier.fit(X_train, y_train)
    accuracies.append(classifier.score(X_test, y_test))
k_list = range(1, 65)
print(accuracies)
```

```python
plt.plot(k_list, accuracies)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Severe Coral Bleaching Classifier')
plt.show


plt.plot(k_list, accuracies)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':np.arange(1,52)}
knn = KNeighborsClassifier()

# This will find the best score and the number of neighbors given a certain cross-validation size
# Let's start with 5 cross-validation sets
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_score_)
print(knn_cv.best_params_)

# Now 10 cross-validation sets
knn_cv= GridSearchCV(knn,param_grid,cv=10)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_score_)
print(knn_cv.best_params_)
```
